```python
import os
import random
import pygame
from pygame import *
from time import sleep
import game_config as gc

# Updated screen size, margin, and image size
SCREEN_SIZE = 600
IMAGE_SIZE = 150
MARGIN = 10
NUM_TILES_SIDE = 4
NUM_TILES_TOTAL = 16  # Updated to match the desired number of tiles

ASSET_DIR = 'assets'
ASSET_FILES = [x for x in os.listdir(ASSET_DIR) if x[-3:].lower() == 'png']

# Initialize counter and font
match_counter = 0
pygame.font.init()  # Initialize the font module
font = pygame.font.SysFont(None, 36) 

animals_count = dict((a, 0) for a in ASSET_FILES)

def available_animals():
    return [animal for animal, count in animals_count.items() if count < 2]

class Animal:
    def __init__(self, index):
        self.index = index
        self.name = random.choice(available_animals())
        self.image_path = os.path.join(gc.ASSET_DIR, self.name)
        self.row = index // gc.NUM_TILES_SIDE
        self.col = index % gc.NUM_TILES_SIDE
        self.skip = False
        self.image = image.load(self.image_path)
        self.image = transform.scale(self.image, (gc.IMAGE_SIZE - 2 * gc.MARGIN, gc.IMAGE_SIZE - 2 * gc.MARGIN))
        self.box = self.image.copy()
        self.box.fill((200, 200, 200))

        animals_count[self.name] += 1

def find_index_from_xy(x, y):
    row = y // gc.IMAGE_SIZE
    col = x // gc.IMAGE_SIZE
    index = row * gc.NUM_TILES_SIDE + col
    return row, col, index

def reset_game(tiles):
    for tile in tiles:
        tile.skip = False
    return []

def game():
    global match_counter
    pygame.init()
    display.set_caption('My Game')
    screen = display.set_mode((SCREEN_SIZE, SCREEN_SIZE))
    matched = image.load('other_assets/matched.png')
    running = True

    # Generate tiles based on the updated NUM_TILES_TOTAL
    tiles = [Animal(i) for i in range(NUM_TILES_TOTAL)]
    current_images_displayed = []

    while running:
        current_events = event.get()

        for e in current_events:
            if e.type == pygame.QUIT:
                running = False

            if e.type == pygame.KEYDOWN:
                if e.key == pygame.K_ESCAPE:
                    running = False
                elif e.key == pygame.K_r:  # Press 'r' to reset the game
                    current_images_displayed = reset_game(tiles)
                    match_counter = 0  # Reset the match counter

            if e.type == pygame.MOUSEBUTTONDOWN:
                mouse_x, mouse_y = pygame.mouse.get_pos()
                row, col, index = find_index_from_xy(mouse_x, mouse_y)
                if index not in current_images_displayed and not tiles[index].skip:
                    if len(current_images_displayed) > 1:
                        current_images_displayed = current_images_displayed[1:] + [index]
                    else:
                        current_images_displayed.append(index)

        # Display animals
        screen.fill((255, 255, 255))

        total_skipped = 0

        for i, tile in enumerate(tiles):
            current_image = tile.image if i in current_images_displayed else tile.box
            if not tile.skip:
                screen.blit(current_image, (tile.col * IMAGE_SIZE + MARGIN, tile.row * IMAGE_SIZE + MARGIN))
            else:
                total_skipped += 1

        # Update and display the match counter
        text = font.render("Matches: " + str(match_counter), False, (0, 0, 0))
        # Adjust the position of the match label based on screen size
        screen.blit(text, (10, SCREEN_SIZE - 30))

        display.flip()

        # Check for matches
        if len(current_images_displayed) == 2:
            idx1, idx2 = current_images_displayed
            if tiles[idx1].name == tiles[idx2].name:
                tiles[idx1].skip = True
                tiles[idx2].skip = True
                match_counter += 1  # Increment the match counter
                # Display matched message
                sleep(0.2)
                screen.blit(matched, (0, 0))
                display.flip()
                sleep(0.5)

                current_images_displayed = []

        if total_skipped == len(tiles):
            running = False

game()
quit()
```

```python
import pygame
import game_config as gc

from pygame import display, event, image, font
from time import sleep
from animal import Animal

def find_index_from_xy(x, y):
    row = y // gc.IMAGE_SIZE
    col = x // gc.IMAGE_SIZE
    index = row * gc.NUM_TILES_SIDE + col
    return row, col, index

def reset_game(tiles):
    for tile in tiles:
        tile.skip = False
    return []

def game():
    pygame.init()
    display.set_caption('My Game')
    
    # Adjusted frame size to accommodate the match label and message
    screen = display.set_mode((gc.SCREEN_SIZE, gc.SCREEN_SIZE + 80))
    
    matched = image.load('other_assets/matched.png')
    running = True
    tiles = [Animal(i) for i in range(0, gc.NUM_TILES_TOTAL)]
    current_images_displayed = []

    # Create fonts for displaying the match counter and message
    font.init()
    match_font = font.Font(None, 36)
    message_font = font.Font(None, 24)

    # Initialize match counter
    match_counter = 0

    matched_indices = []

    while running:
        current_events = event.get()

        for e in current_events:
            if e.type == pygame.QUIT:
                running = False

            if e.type == pygame.KEYDOWN:
                if e.key == pygame.K_ESCAPE:
                    running = False
                    
            if e.type == pygame.KEYDOWN:
                if e.key == pygame.K_ESCAPE:
                    running = False
                elif e.key == pygame.K_r:  # Press 'r' to reset the game
                    current_images_displayed = reset_game(tiles)
                    match_counter = 0  # Reset the match counter

            if e.type == pygame.MOUSEBUTTONDOWN:
                mouse_x, mouse_y = pygame.mouse.get_pos()
                row, col, index = find_index_from_xy(mouse_x, mouse_y)
                if index not in current_images_displayed and index not in matched_indices:
                    if len(current_images_displayed) > 1:
                        current_images_displayed = current_images_displayed[1:] + [index]
                    else:
                        current_images_displayed.append(index)

        # Display animals
        screen.fill((255, 255, 255))

        total_skipped = 0

        for i, tile in enumerate(tiles):
            current_image = tile.image if i in current_images_displayed else tile.box
            if not tile.skip:
                screen.blit(current_image, (tile.col * gc.IMAGE_SIZE + gc.MARGIN, tile.row * gc.IMAGE_SIZE + gc.MARGIN))
            else:
                total_skipped += 1

        # Check for matches
        if len(current_images_displayed) == 2:
            idx1, idx2 = current_images_displayed
            if tiles[idx1].name == tiles[idx2].name:
                # Display the matched image for a short duration
                screen.blit(matched, (gc.SCREEN_SIZE // 2 - matched.get_width() // 2, gc.SCREEN_SIZE // 2 - matched.get_height() // 2))
                display.flip()
                sleep(0.5)

                tiles[idx1].skip = True
                tiles[idx2].skip = True
                match_counter += 1  # Increment the match counter
                matched_indices.extend([idx1, idx2])

                current_images_displayed = []

        # Display the match label at the bottom
        match_label = match_font.render("Matches: ", True, (0, 0, 0))
        match_value = match_font.render(str(match_counter), True, (0, 0, 0))
        
        # Display the message label at the bottom
        message_label = message_font.render("Press 'R' to restart", True, (0, 0, 0))
        
        screen.blit(match_label, (10, gc.SCREEN_SIZE))
        screen.blit(match_value, (match_label.get_width() + 20, gc.SCREEN_SIZE))
        
        # Display the message label below the match label
        screen.blit(message_label, (10, gc.SCREEN_SIZE + 10))

        display.flip()

        if total_skipped == len(tiles):
            running = False

    # Allow matched tiles to be displayed for a moment before quitting
    sleep(2)

game()
quit()
```
